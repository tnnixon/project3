```python
import requests
import json
import prettytable
import pprint
headers = {'Content-type': 'application/json'}
data = json.dumps({"seriesid": ['CUSR0000SA0'],"startyear":"2020", "endyear":"2022"})

p = requests.post('https://api.bls.gov/publicAPI/v2/timeseries/data/', data=data, headers=headers)
json_data = json.loads(p.text)
pprint.pprint(json_data)
print("received json data")
for series in json_data['Results']['series']:
    x=prettytable.PrettyTable(["series id","year","period","periodName","value"])
    seriesId = series['seriesID']
    for item in series['data']:
        year = item['year']
        period = item['period']
        periodName=item['periodName']
        value = item['value']
        print(item['year'],",",item['periodName'],",",item['value'])
#        footnotes=""
#        for footnote in item['footnotes']:
#            if footnote:
#                footnotes = footnotes + footnote['text'] + ','
        if 'M01' <= period <= 'M12':
            x.add_row([seriesId,year,period,periodName,value])
    output = open(seriesId + '.txt','w')
    output.write (x.get_string())
    output.close()
    
```
